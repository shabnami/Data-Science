```python
normalized_data = json_normalize(data, 'mjtheme_namecode')
normalized_data.code.value_counts().head(10)
```

## In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
missing_data = normalized_data[normalized_data.name == '']
#missing_data
```

```python
missing_data.head()
```

```python
#creating a dictionary based on the values of the code
code_dict = {}
for i in normalized_data.code.unique():
    code_dict[i] = normalized_data[normalized_data.code==i].name.iloc[0]
code_dict
```

```python
code_dict['11']='Environment and natural resources management'
code_dict
```

```python
missing=normalized_data[['code','name']].drop_duplicates().query("name != ''")
filled_missing = normalized_data.merge(missing,on='code')[['code','name_y']]
filled_missing.columns = ['code','name']
#filled_missing
```

```python
missing_data = filled_missing[filled_missing.name == '']
missing_data
```
